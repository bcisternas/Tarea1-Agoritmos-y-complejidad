```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Leer el archivo CSV utilizando delimitaci√≥n por espacio en blanco
df_rectangulares = pd.read_csv('resultados_matrices_rectangulares.csv', delim_whitespace=True)

# Mostrar las primeras filas del DataFrame para verificar que los datos se leyeron correctamente
print(df_rectangulares.head())

# Extraer los datos del CSV
tamanios_rectangulares = df_rectangulares['tama√±o']
algoritmo_original_rectangulares = df_rectangulares['algoritmo_original']
algoritmo_optimizado_rectangulares = df_rectangulares['algoritmo_optimizado']

# Crear las curvas te√≥ricas m^3 y n log n (adaptadas para las matrices rectangulares)
n = np.array(tamanios_rectangulares)
m_cubed = n ** 3

# Crear el gr√°fico para matrices rectangulares
plt.figure(figsize=(10, 6))
plt.plot(tamanios_rectangulares, algoritmo_original_rectangulares, label='Algoritmo Original', marker='o')
plt.plot(tamanios_rectangulares, algoritmo_optimizado_rectangulares, label='Algoritmo Optimizado', marker='o')

# Graficar las curvas te√≥ricas m^3 y n log n
plt.plot(n, m_cubed / max(m_cubed) * max(algoritmo_original_rectangulares), label='$n^3$', linestyle='--', color='purple')

# Configuraci√≥n del gr√°fico
plt.title('Comparaci√≥n de Algoritmos de Multiplicaci√≥n de Matrices Rectangulares')
plt.xlabel('Tama√±o de la Matriz')
plt.ylabel('Tiempo de Ejecuci√≥n (ms)')
plt.legend()
plt.grid(True)

# Guardar y mostrar el gr√°fico
plt.savefig('graficos/matrices_rectangulares_comparacion.png')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Leer el archivo CSV utilizando delimitaci√≥n por espacio en blanco
df_cuadradas = pd.read_csv('resultados_matrices_cuadradas.csv', delim_whitespace=True)

# Mostrar las primeras filas del DataFrame para verificar que los datos se leyeron correctamente
print(df_cuadradas.head())

# Extraer los datos del CSV
tamanios_cuadradas = df_cuadradas['tama√±o']
algoritmo_original_cuadradas = df_cuadradas['algoritmo_original']
algoritmo_optimizado_cuadradas = df_cuadradas['algoritmo_optimizado']
algoritmo_strassen_cuadradas = df_cuadradas['algoritmo_strassen']

# Crear las curvas te√≥ricas n^2 y n log n
n = np.array(tamanios_cuadradas)
m_cubed = n ** 3

# Crear el gr√°fico para matrices cuadradas
plt.figure(figsize=(10, 6))
plt.plot(tamanios_cuadradas, algoritmo_original_cuadradas, label='Algoritmo Original', marker='o')
plt.plot(tamanios_cuadradas, algoritmo_optimizado_cuadradas, label='Algoritmo Optimizado', marker='o')
plt.plot(tamanios_cuadradas, algoritmo_strassen_cuadradas, label='Algoritmo Strassen', marker='o')

# Graficar las curvas te√≥ricas
plt.plot(n, m_cubed / max(m_cubed) * max(algoritmo_original_cuadradas), label='$n^3$', linestyle='--', color='purple')

# Configuraci√≥n del gr√°fico
plt.title('Comparaci√≥n de Algoritmos de Multiplicaci√≥n de Matrices Cuadradas')
plt.xlabel('Tama√±o de la Matriz')
plt.ylabel('Tiempo de Ejecuci√≥n (ms)')
plt.legend()
plt.grid(True)

# Guardar y mostrar el gr√°fico
plt.savefig('graficos/matrices_cuadradas_comparacion.png')
plt.show()
```
